```python
#  matplotlib基本作圖
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei"]
days=range(1,7)
celsius=[25.6,23.2,18.5,28.3,26.5,30.5]
plt.plot(days,celsius,"ko-")   # 先x軸，後y軸，後面可以接參數:顏色、線條、圖形，順序可以亂倒!!!
# 顏色(r, y, g, b, c, m, k, w)紅黃綠藍青紫黑白
# 線條(-   --   .  -:)實線 短虛線 點虛線 短虛線+點虛線
# 點的圖形( . , o ^ s) 點、無、大點、三角形、四方型
plt.grid(True)   #  會多了背景的網格!!!
plt.xlabel("Days")
plt.ylabel("Celsius")
plt.title("下周溫度")
print(plt.axis())    #  可以知道x軸最小值、x軸最大值、y軸最小值、y軸最大值
plt.axis([0.5,6.5,16,32])   #  然後再依照想要的x/y軸最大最小值做修改
plt.show()
```

```python
#  上課的demo資料
import numpy as np
import matplotlib.pyplot as plt

#  把那些紅紅的注意事項消掉
import warnings
warnings.filterwarnings('ignore')

labels = ["Python","C++","Java","JS","C","C#"]
index = np.arange(len(labels))
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
change = [1.12,0.3,-1.69,0.29,3.41,-0.45]
```

```python
#  繪製長條圖plt.bar(x軸,y軸)
plt.bar(index,ratings)    # 此處x軸用index，若用labels可以直接省略下一行
plt.xticks(index,labels)  # 每一欄要給他什麼標籤
plt.ylabel("使用率")
plt.title("程式設計使用率")  
plt.show()
```

```python
#  繪製水平長條圖，改成horizontal的：bar+h !!!!
plt.barh(index,ratings)     #  變成barh(y軸, x軸)了!!!!
plt.yticks(index,labels)    #  改成y軸
plt.xlabel("使用率")        #  改成x軸 
plt.title("程式設計使用率") 
plt.show()
```

```python
#  繪製直方圖histogram
x = [21,42,23,4,5,26,77,88,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5   #  切割成幾份
n, bins, patches = plt.hist(x,num_bins)
print(n)
print(bins)
print(patches)
plt.show()
```

```python
#  繪製派圖pie chart
labels = ["Python","C++","Java","JS","C","C#"]
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
plt.pie(ratings,labels=labels)
plt.title("程式語言使用率")
plt.axis("equal")   # 不知道這行是什麼??，好像有平移
plt.show()
```

```python
#  繪製子圖表subplots
x = np.linspace(0,10,50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.subplot(2,1,1)    # subplot(row的數量,column的數量,第幾個row)
plt.plot(x,sinus,"g-o")
plt.subplot(2,1,2)
plt.plot(x,cosinus,"b--s")
plt.show()
```

```python
#  繪製多軸圖表，畫在同一個表中，這邊示範共用x軸，但用不同的y軸
x = np.linspace(0,10,50)
sinus = np.sin(x)
sinhs = np.sinh(x)
fig,bx = plt.subplots()  # bx可以隨便改
bx.plot(x,sinus,'r-o')   #第一條x
bx2 = bx.twinx()     #第二條x，共用x軸
bx2.plot(x,sinhs,'g--')   #第二條x
plt.show()
```
