```python
#  載入資料包
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline  # 加上這行的話可以直接秀圖
import warnings
warnings.filterwarnings('ignore')
                        
current_palette = sns.color_palette()   #調色盤
sns.palplot(current_palette)   #  用來看現在使用的色表是哪一種
```

```python
x=np.arange(1,9)
y=[4,2,6,1,6,3,7,8]
plt.bar(x,y)
sns.set_context("paper")
sns.set_style("white")
sns.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=5, trim=False)
plt.show()
```

```python
#  seaborn提供的繪圖風格
sns.set(context="notebook", style="darkgrid",palette="deep",font="sans-serif", font_scale=1,color_codes=True, rc=None)  #  palette調色盤、字形型態、字體大小
#  移除上方、右方axis:  sns.despine()，offset是偏移量填純量，right和bottom會消失所以是True
sns.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)
#  避免subplot之間重疊:  plt.tight_layout()
#  四種context:    影響標籤和線條，默認使用notebook
sns.set_context("notebook")  #  default, font=1
sns.set_context("paper")   #  font=0.8
sns.set_context("talk")    #  font=1.3
sns.set_context("poster")  #  font=1.6
#  五種style
sns.set_style("darkgrid")     # 深色網格
sns.set_style("whitegrid")    # 白色網格
sns.set_style("dark")         # 深色背景
sns.set_style("white")        # 白色背景
sns.set_style("ticks")        # 白色背景+ticks
#  六種palette
sns.set_palette("deep")
sns.set_palette("muted")
sns.set_palette("pastel")  # 最透明
sns.set_palette("bright")  # 最亮
sns.set_palette("dark")    # 最深
sns.set_palette("colorblind")
```

```python
#  seaborn提供的6種調色板 
theme_list=['deep', 'muted', 'pastel','bright', 'dark', 'colorblind'] 
for i in theme_list: 
    sns.palplot(sns.color_palette(i))   #  用來看現在使用的色表是哪一種
```

```python
#  設置調色盤，基本的sns.color_palette("hls",數目) 或 sns.hls_palette(數目)
sns.color_palette(palette=None, n_colors=None, desat=None)  # 有三個參數可以設定
sns.color_palette("hls", 25)   #  (色彩空間名稱,顏色數量)
sns.hls_palette(25)
```

```python
#  把設定的調色盤代入資料

data = np.random.normal(size=(20, 12)) + np.arange(12) / 2 
sns.boxplot(data=data,palette=sns.color_palette("hls", 12))
```

```python
# hls=hue + saturation + lightness，從RGB經過數學轉換而來，h,l,s可以填0到1的數字
sns.hls_palette(n_colors=6, h=0.01,l=0.6, s=0.65)
```

```python
#使用默認亮度和飽和度 
sns.palplot(sns.hls_palette(3))
```

```python
sns.palplot(sns.hls_palette(3, l=0.8, s=0.8))
```

```python
#  直接要某某顏色，用sns.xkcd_rgb("顏色名")
sns.xkcd_rgb["cyan"]
plt.plot([0, 1], [0, 1], sns.xkcd_rgb["purple"], lw=2)
plt.plot([0, 1], [0, 2], sns.xkcd_rgb["green"], lw=3)
plt.plot([0, 1], [0, 3], sns.xkcd_rgb["orange"], lw=4)   # lw表示線寬
```

```python
#  想要用連續性色版的話，sns.colot_plaette(顏色名)
sns.palplot(sns.color_palette("Blues")) 
sns.palplot(sns.color_palette("Reds")) 
sns.palplot(sns.color_palette("Greens_r"))    # 想要由深到淺的話加上_r
```

```python
#  色調線性變換 sns.color_palette("cubehelix") 或是 sns.cubegelix_palette(數目)
sns.palplot(sns.color_palette()) 
sns.palplot(sns.color_palette("cubehelix"))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=3, rot=0.5))
```

```python
#  訂製連續調色盤  sns.light_palette(顏色名), sns.dark_palette(顏色名)
sns.palplot(sns.light_palette("green")) 
sns.palplot(sns.dark_palette("green"))
```

```python
#  類似等高線圖，但老師可以我不行
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T
pal = sns.dark_palette("green", as_cmap=True) 
sns.kdeplot(x, y, cmap=pal);
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T 
pal = sns.cubehelix_palette(light=1, as_cmap=True) 
sns.kdeplot( x, y, cmap=pal, shade=True)
```
