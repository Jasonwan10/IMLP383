#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，

#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
#隨機數1-100
import random
num=random.randint(1,100)

#猜測條件判斷
lst=[0,num,100]    #猜數字的list
condition=1  #繼續玩的條件
times=1      #玩的次數
while condition==1:
    print("猜第"+str(times)+"次")
    digit=1      #限制輸入整數
    while digit==1:
        guess=input("選一個整數吧:")
        if guess.isdigit()==1:
            guess=int(guess)
            digit=0
        else:
            print("請輸入整數!!")
    lst.append(guess)     
    lst.sort()
    num_index=lst.index(num)    #num在list中的index
    num_plus=lst[num_index+1]   #num後一位數
    num_minus=lst[num_index-1]  #num前一位數
    if guess==num:
        print("賓果!!!猜對了")
        break
    elif guess>=num_plus:
        print("再小一點!!")
        print("猜數字範圍"+str(num_minus)+"<?<"+str(num_plus))
    elif guess<=num_minus:
        print("再大一點!!")
        print("猜數字範圍"+str(num_minus)+"<?<"+str(num_plus))
    times+=1
    while 1:
        #只能玩8次
        if times==9:
            print("玩超過8次了,88")
            print("答案是"+ str(num))
            condition=0
            break
        ask=input("還想要繼續玩?(y/n)")
        if ask=="n":
                condition=0  
                break
        if ask=="y":
                print("===========================================")  
                break
        
````

* Hint:差集(減法)、交集


````python
m={'柯南','灰原','步美','美環','光彦'}
e={'柯南','灰原','丸尾','野口','步美'}

a=m.intersection(e)
b=m.difference(a)
c=e.difference(a)

a1=list(a)
b1=list(b)
c1=list(c)

a1.sort()
b1.sort()
c1.sort()

print("數學及格但英文不及格") 
print(b1) 

print("英文及格但數學不及格")
print(c1) 

print("兩個都及格")
print(a1)    

print("全班")
print(m.union(e))
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
def convertInputString():
    #請使用者輸入一串字串，先轉換字串為統一小寫
    rawInput = input("\nPlease enter a word, phrase, or a sentence to check if it is a palindrome: ") 
    rawstring=rawInput.lower()
    rawlist=list(rawstring)
    return rawlist

def stripAnalphabetics(dirtyList): 
    analphabeticList = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
    for i in analphabeticList:
        if i in dirtyList:
            dirtyList.remove(i)
            return stripAnalphabetics(dirtyList)
    return dirtyList

def runPalindromeCheck(straightList):
    reverseList = straightList[::-1] 
    if reverseList==straightList: 
        return "The text you have entered is a palindrome!" 
    else: 
        return "The text you have entered is not a palindrome." 

def main(): 
    print("\nPalindrome checker")
    originalList=convertInputString()
    originalList=stripAnalphabetics(originalList)
    result=runPalindromeCheck( originalList)
    return result

main() 
````

````python
fin_E = open("data/english_list.csv","r",encoding='UTF-8')
fin_M = open("data/math_list.csv","r",encoding='UTF-8')
lisE=[]
lisM=[]
name=[]
for line in fin_E.readlines():
#     print(line,end="")   #每print一次會換行，end=""可以把換行去掉
    line = line.strip().split(",")  #strip去除空格、split(",")是看到逗號就切割
#     print(line)
    name.append(line[0])
    lisE.append(line[1])
    
for line in fin_M.readlines():
    line = line.strip().split(",")
    lisM.append(line[1])


score=[]
fout = open("Score.csv","w")
line=''
for i in range(1, len(name)): #第0是"姓名"、"英文"、"數學"
    score.append(int(lisE[i])+int(lisM[i]))
    list1 = [name[i],str(score[i-1]),"\n"] #補\n為了最後寫入csv檔的時候會換行
    print(list1) 
    line = ",".join(list1)   #把list變回str，用於csv: 名字,分數,(換行)名字,分數,(換行)名字,分數....
    print(line)
    fout.write(line)
fin_E.close()
fin_M.close()
fout.close()

'''
line=''
fout = open("data/Score.csv","r")
for i in (len(name)):
    print()
'''
````

````python
# split和join解說
s="1 2 3"   # 只有字串有split
s=s.split()   #代表用空格把它隔開
print(s)
print(type(s))  #是list

s="".join(s)   #不太直觀， 公式：把"指定的東西"合在一起.join(要被合在一起的序列)
print(s)
print(type(s))  #是str
````
